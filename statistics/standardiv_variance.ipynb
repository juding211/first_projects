```python
**Variance**
A numerical measure of how much the data points spread out from the mean.

Mean:(70+72+68+74+66)/5=70

Differences from the mean:[0, 2, -2, 4, -4]

Squared differences:[0, 4, 4, 16, 16]

Sum of squared differences:40

Sample variance:40/5−1=10


So, the variance of this data is 10.
```

분산
데이터가 평균으로부터 얼마나 퍼져 있는지를 수치로 나타낸 것

데이터: lifeExp = [70, 72, 68, 74, 66]

평균: (70+72+68+74+66)/5=70

각 값에서 평균을 뺀 값: [0, 2, -2, 4, -4]

제곱: [0, 4, 4, 16, 16]

합계: 40

분산 (표본 분산): 40/5−1=10

→ 이 데이터의 분산은 10입니다.

```python
df['lifeExp'].std() #standard deviation 표준편차
```

**표준편차: 분산(Variance)의 제곱근(sqrt)**

→ 데이터가 평균에서 얼마나 떨어져 있는지를 원래 단위로 알려줘서 더 해석이 쉬워진다.

평균: (70+72+68+74+66)/5=70

분산 (표본 분산): 40/5−1=10

s = 10루트 ≈3.16
