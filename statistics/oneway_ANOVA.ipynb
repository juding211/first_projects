```python
import pandas as pd #Library for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series.

df = pd.read_csv('일원분산분석_실습.csv', sep=',', encoding='CP949')
df
```

```python
import scipy as sp
import numpy as np
import pingouin as pg
df1 = df[['store', 'satisfaction']]
store = []
for i in range(1, 4):
    store.append(df1[df1['store']==i]['satisfaction'].values)
sp.stats.levene(store[0], store[1], store[2])  #일원분산분석 전에 등분산성을 해야함. 여기선 levene의 등분산성 사용. pvalue0.2여서 귀무가설 채택. 차이가 없다=등분산성 이루고 있다.
```

```python
# One-way ANOVA
print(pg.anova(dv='satisfaction', between='store', data=df1))
```

pvalue 0,00009 대립가설 채택.  store 1,2,3, 만족도 값에 차이가 있다.

등분산성이 만족하지 않을 경우 welch anova라는 코드를 사용

print(pg.welch_anova(dv='satisfaction', between='store', data=df1))

위에는 차이가 있다는 것만 알 뿐, 구체적으로 어떤 집단과 어떤 집단의 만족도에 차이가 있는지를 모름.  그래서 **사후 분석**이라는 것을 해야한다.

```python
!pip install scikit-posthocs
```

**scikit-posthocs** is a Python library for performing multiple comparisons (post-hoc tests) after statistical tests like ANOVA or Kruskal-Wallis.
It is especially useful when you want to find which specific groups are different from each other.

```python
import scikit_posthocs

df1['store'] = df1['store'].astype(str)
print(scikit_posthocs.posthoc_scheffe(df1, val_col='satisfaction', group_col='store'))
```

A와 B store은 차이가 없다. pvalue->0.15
A와 C store은 차이가 있다. pvalue->0.0028
B와 C store이 차이가 있다. pvalue->0.000013
