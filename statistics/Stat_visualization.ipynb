```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

df = pd.read_csv('gapminder.tsv', sep='\t')
df
```

```python
# Histogram

plt.hist(df['lifeExp'], alpha = 0.4, bins = 7, rwidth=2, color='green')
```

```python
# Pie Chart

df2= df.groupby('continent')
df2.describe()
```

```python
sizes = [624, 300, 396, 360, 24]
labels = ['Africa', 'Americas', 'Asia', 'Europe', 'Oceania']
colors = ['red', 'green', 'blue', 'yellow', 'purple']
plt.pie(sizes, labels = labels, colors = colors, autopct = '%.1f%%', shadow = False, startangle = 90 )
plt.axis('equal')
plt.show()
```

autopct = 포매팅 소수점 첫 번째까지 '%.1f%%'

```python
# Boxplot
import seaborn as sns

sns.boxplot(x='continent', y='lifeExp', data=df)
```
