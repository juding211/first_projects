# Relation Algebra

*   Relational Algebra: Procedural theory behind how SQL is executed inside DBMS






# . Relational Algebra → SQL Mapping
Algebra
σ (Selection)	WHERE	Filter rows

π (Projection)	SELECT	Select columns

⋈ (Join)	JOIN ... ON	Combine two tables

× (Product)	FROM A, B	Cartesian product

∪ (Union)	UNION	Combine without duplicates

− (Difference)	EXCEPT	Subtract one result from another

∩ (Intersection)	INTERSECT	Keep common rows

1. Why does it matter to know relational algebra/calculus for SQL writing?

SQL is declarative, but DBMS uses relational algebra/calculus internally

Knowing this will help me to:

1)   Write more logical and efficient queries.

2)   understand how queries are optimized.

3)   Predict and improve execution plans.



4. Execution Order & Optimization
Relational Algebra operations have order

Changing the order can affect performance

Example:

Doing JOIN → filter may be slower than

filter → JOIN (because the dataset gets smaller before the JOIN)

Therefore, learning the theory helps you write smarter queries
