```python
import pingouin as pg
import scikit_posthocs

df1 = df[['store', 'area', 'satisfaction']]
print(pg.anova(dv='satisfaction', between=['store', 'area'], data=df1))
```

In this two-way ANOVA analysis, we examined whether two factors—store and area—have an effect on satisfaction, and whether there is an interaction effect between them. Here are the results explained in simple terms:

Effect of store:
The statistical test shows an F-value of 12.34 with a p-value of 0.00029.
Since this p-value is less than 0.05 (written as p < 0.05), the result is considered statistically significant.
-> This means that satisfaction levels differ significantly depending on the store.

Effect of area:
The F-value is 29.99 and the p-value is 0.0000007011, which is also much smaller than 0.05.
-> This indicates that area has a strong and statistically significant effect on satisfaction.

Interaction between store and area:
The F-value here is 1.33 and the p-value is 0.2927, which is greater than 0.05 (written as p > 0.05).
-> This result is not statistically significant, meaning that the effect of store on satisfaction does not depend on the area, and vice versa.



```python
df1['store'].astype(str)
print(scikit_posthocs.posthoc_scheffe(df1, val_col='satisfaction', group_col='store'))
```

2,3번 브랜드의 만족도에 차이가 있다.

```python
df1['area'].astype(str)
print(scikit_posthocs.posthoc_scheffe(df1, val_col='satisfaction', group_col='area'))
```
